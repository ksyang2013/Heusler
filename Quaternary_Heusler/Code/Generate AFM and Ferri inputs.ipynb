```python
import os
from main.xstructure import *
import numpy as np
import shutil

mag = ['Co','Cr','Fe','Mn','Mo','Ni','V','W','Ce','Eu']
dist = '.' 
for files in os.listdir(dist):
    file_path = os.path.join(dist,files)
    if os.path.isdir(file_path)==True and "ipynb" not in file_path:
        os.system("mkdir ./%s/FM"%file_path)
        str_ps = file2string("%s/POSCAR"%file_path)
        poscar_1 = xstructure(str_ps)
        fm_ = list(set(mag)&set(poscar_1.species))
        num = len(fm_)
        if num == 0 or num == 1:
            os.system("rm  -r ./%s"%file_path)
            continue
        if num == 2:
            magmom = []
            for i in range(0,poscar_1.natm):
                if poscar_1.atoms[i].name == fm_[0]:
                    magmom.append(int(5))
                elif poscar_1.atoms[i].name == fm_[1]:
                    magmom.append(int(-5))
                else:
                    magmom.append(int(0))

        if num == 3:
            magmom = []
            for i in range(0,poscar_1.natm):
                if poscar_1.atoms[i].name == fm_[0]:
                    magmom.append(int(5))
                elif poscar_1.atoms[i].name == fm_[1]:
                    magmom.append(int(-5))
                elif poscar_1.atoms[i].name == fm_[2]:
                    magmom.append(int(5))
                else:
                    magmom.append(int(0))    

        if num == 4:
            magmom = []
            for i in range(0,poscar_1.natm):
                if poscar_1.atoms[i].name == fm_[0]:
                    magmom.append(int(5))
                elif poscar_1.atoms[i].name == fm_[1]:
                    magmom.append(int(-5))
                elif poscar_1.atoms[i].name == fm_[2]:
                    magmom.append(int(5))
                elif poscar_1.atoms[i].name == fm_[3]:
                    magmom.append(int(-5))
                else:
                    magmom.append(int(0))

    
        for f in os.listdir(file_path):
            fp = os.path.join(file_path,f)
            shutil.move("%s"%fp, "%s/FM"%file_path)    
        shutil.copytree("%s/FM"%file_path, "%s/Ferri"%file_path)
        All_lines = ""
        with open ("%s/Ferri/INCAR"%file_path) as ff: 
            data = ff.readlines()
            for line in data:
                if "MAGMOM" in line:
                    m = ""
                    for mm in magmom:
                        m = m + " "+ str(mm)
                    All_lines = All_lines + "MAGMOM = " + m + "\n"
                else:
                    All_lines = All_lines+line
        string2file(All_lines,"%s/Ferri/INCAR"%file_path)
        print(file_path,"\n",magmom)
```
